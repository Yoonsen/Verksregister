## create a triple, worksid, oldfiles (before 1800) and new files (after 1900)

A triple consists of the works ID plus as list of old works together with a list of newer works. This can be made to any arbitrary granularity.

```python
triple = {}
for i in result:
    data = result[i]
    old = [x for x in data if int(data[x][0]) <= 1900 ]
    new = [x for x in data if int(data[x][0]) > 1900 ] 
    if old != [] and new != []:
        triple[i] = (old, new)
```
